```python

# Setting up the workspace by importing all of the necessary libraries.
#from importlib import reload

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
def fix_placeholder(df=None, column=None ,placeholder=None, value=0):
    for item in placeholder:
        df.loc[df[column] == item, column]=value
```

#### Importing the data and combining two datasets
- dietdf contains information collected about each respondents dietary intake
- quesdf contains information obtained from a survey conducted on the respondents

We are joining both dataframes together using SEQN which will be renamed to id.

```python
dietdf = pd.read_csv('diet.csv')
dietdf.head()
```

```python
display(dietdf.head())
print(dietdf.shape)
```

```python
dietdf.describe()
```

```python
quesdf = pd.read_csv('questionnaire.csv')
quesdf.head()
```

#### This Data Set has only a few items we need, so we will isolate those for our analysis

```python
quesdf = quesdf[['SEQN', 'DIQ010', 'DIQ160']]
quesdf.head()
```

#### In preparation for joining, we will rename the index of one of the DF's

```python
quesdf.rename({'SEQN':'SEQN1'}, axis=1, inplace=True);
quesdf.head()
```

#### Many of the columns are not necessary for our analysis, so we will drop them here.
- 'SEQN1' is the same as 'SEQN' and is not necessary after the join
- 'DR1EXMER' is an identifier for the interviewer
- 'DR1DBIH' is the number of days elapsed between the exam and interview
- 'DR1LANG' is the language used by the respondent for the interview
- 'DR1MNRSP' is the person who answered the questions, subject or caretaker
- 'DR1HELPD' identifies who helped answer the questions, if anyone

```python
df = pd.merge(dietdf, quesdf, left_on='SEQN', right_on='SEQN1', how='left')
df.drop('SEQN1', axis=1, inplace=True)
df.drop('DR1EXMER', axis=1, inplace=True)
df.drop('DR1DBIH', axis=1, inplace=True)
df.drop('DR1LANG', axis=1, inplace=True)
df.drop('DR1MNRSP', axis=1, inplace=True)
df.drop('DR1HELPD', axis=1, inplace=True)
df.rename(columns={'SEQN':'id'}, inplace=True)
df.head()
```

```python
df.describe()
```

#### There are numerous NaN's, so we will deal with them here

```python
nan_list_1 = [
    'DRABF',
    'DRDINT',
    'DR1DAY',
    'DBQ095Z',
    'DBD100',
    'DRQSPREP',
    'DR1STY',
    'DR1SKY',
    'DRQSDIET',
    'DRQSDT1',
    'DRQSDT2',
    'DRQSDT3',
    'DRQSDT4',
    'DRQSDT5', 
    'DRQSDT6',
    'DRQSDT7',
    'DRQSDT8',
    'DRQSDT9',
    'DRQSDT10',
    'DRQSDT11',
    'DRQSDT12',
    'DRQSDT91',
    'DR1TNUMF'
]
```

```python
for item in nan_list_1:
    df.loc[df[item] > 91] = 0
```

```python
df.fillna(0,inplace=True);
```

```python
for col in df.columns:
    print(f'\n{col}')
    print(df[col].unique())
```
